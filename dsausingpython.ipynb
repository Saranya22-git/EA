```python
arr = [10, 20, 40, 50]
n = 4        # current size
pos = 2      # insert position
value = 30

arr.append(0)   # increase size

for i in range(n-1, pos-1, -1):
    arr[i+1] = arr[i]

arr[pos] = value

print(arr)
```
